```python
df=pd.read_csv("transaction_data.csv")
```

```python
df
```

```python
df.info()
```

```python
df.drop_duplicates(inplace=True)
```

```python
# 1. Total spending per customer
total_spending = df.groupby('LYLTY_CARD_NBR')['TOT_SALES'].sum().reset_index()
total_spending.rename(columns={'TOT_SALES': 'TOTAL_SPENDING'}, inplace=True)
```

```python
total_spending
```

```python
# 2. Total number of transactions per customer
transaction_count = df.groupby('LYLTY_CARD_NBR')['TXN_ID'].count().reset_index()
transaction_count.rename(columns={'TXN_ID': 'TOTAL_TRANSACTIONS'}, inplace=True)
```

```python
transaction_count
```

```python
# 3. Average transaction value per customer
customer_metrics = pd.merge(total_spending, transaction_count, on='LYLTY_CARD_NBR')
customer_metrics['AVG_SALES'] = customer_metrics['TOTAL_SPENDING'] / customer_metrics['TOTAL_TRANSACTIONS']
customer_metrics
```

```python
# 4. Product-level profitability
total_sales_per_product = df.groupby('PROD_NBR')['TOT_SALES'].sum().reset_index()
total_sales_per_product.rename(columns={'TOT_SALES': 'TOTAL_PRODUCT_SALES'}, inplace=True)
```

```python
df
```

```python
# 5. Customer segmentation
df_customer_segmentation = df1[['LYLTY_CARD_NBR', 'LIFESTAGE', 'PREMIUM_CUSTOMER']].drop_duplicates()
```

```python
# Merge customer segmentation with customer metrics
customer_analysis = pd.merge(customer_metrics, df_customer_segmentation, on='LYLTY_CARD_NBR', how='left')
```

```python
# Save the feature engineered dataset
customer_analysis.to_csv("customer_analysis.csv", index=False)
total_sales_per_product.to_csv("product_sales.csv", index=False)
```
